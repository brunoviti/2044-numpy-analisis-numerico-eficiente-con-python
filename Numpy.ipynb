```python
# prompt: ok, necesito la pendiente de la recta de ese grafico

import matplotlib.pyplot as plt
import numpy as np
# Para calcular la pendiente de la recta, podemos usar la función `polyfit` de numpy.
# Esta función ajusta un polinomio a los puntos de datos, y para una línea recta,
# ajustamos un polinomio de grado 1.
# La pendiente es el coeficiente del término de grado 1.

# Ajustar una recta (polinomio de grado 1) a los datos de Moscú
# `fechas` son los valores del eje x y `Moscu` son los valores del eje y
pendiente, intercepcion = np.polyfit(fechas, Moscu, 1)

print(f"La pendiente de la recta de los precios de Moscú es: {pendiente:.4f}")

# Opcionalmente, puedes trazar la recta ajustada en el gráfico
recta_ajustada = pendiente * fechas + intercepcion

plt.figure(figsize=(10, 5))
plt.plot(fechas, Moscu, label='Moscu', marker='o')
plt.plot(fechas, recta_ajustada, color='red', linestyle='--', label=f'Recta Ajustada (Pendiente={pendiente:.4f})')

plt.xlabel('Fecha (Meses desde 2013)')
plt.ylabel('Precio')
plt.title('Precio de las Manzanas en Moscú a lo Largo del Tiempo con Recta Ajustada')

if 'años_unicos' in locals() and 'fechas_años_unicos' in locals():
    plt.xticks(fechas_años_unicos, años_unicos)

plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: necesito hacer lo mismo para todas las ciudades

import matplotlib.pyplot as plt
import numpy as np
# Función para procesar y graficar datos para una ciudad dada
def process_city_data(city_data, city_name, fechas, años_unicos, fechas_años_unicos):
    # Encontrar y manejar valores NaN
    if np.sum(np.isnan(city_data)) > 0:
        print(f"Advertencia: Se encontraron valores NaN en los datos de {city_name}. Realizando interpolación lineal.")
        # Interpolar valores NaN
        nan_indices = np.isnan(city_data)
        city_data[nan_indices] = np.interp(np.where(nan_indices)[0], np.where(~nan_indices)[0], city_data[~nan_indices])

    # Calcular la media
    mean_price = np.mean(city_data)
    print(f"La media del precio de las manzanas en {city_name} es: {mean_price:.2f}")

    # Trazar la recta para el valor de la fruta en la ciudad
    plt.figure(figsize=(10, 5))
    plt.plot(fechas, city_data, label=city_name, marker='o')

    plt.xlabel('Fecha (Meses desde 2013)')
    plt.ylabel('Precio')
    plt.title(f'Precio de las Manzanas en {city_name} a lo Largo del Tiempo')

    if 'años_unicos' in locals() and 'fechas_años_unicos' in locals():
        plt.xticks(fechas_años_unicos, años_unicos)

    plt.legend()
    plt.grid(True)
    plt.show()

    # Calcular la pendiente de la recta ajustada
    pendiente, intercepcion = np.polyfit(fechas, city_data, 1)
    print(f"La pendiente de la recta de los precios de {city_name} es: {pendiente:.4f}")

    # Trazar la recta ajustada
    recta_ajustada = pendiente * fechas + intercepcion

    plt.figure(figsize=(10, 5))
    plt.plot(fechas, city_data, label=city_name, marker='o')
    plt.plot(fechas, recta_ajustada, color='red', linestyle='--', label=f'Recta Ajustada (Pendiente={pendiente:.4f})')

    plt.xlabel('Fecha (Meses desde 2013)')
    plt.ylabel('Precio')
    plt.title(f'Precio de las Manzanas en {city_name} a lo Largo del Tiempo con Recta Ajustada')

    if 'años_unicos' in locals() and 'fechas_años_unicos' in locals():
        plt.xticks(fechas_años_unicos, años_unicos)

    plt.legend()
    plt.grid(True)
    plt.show()

# Lista de datos de precios por ciudad (asumiendo que ya están definidos)
ciudades_data = [Moscu, Kaliningrado, Petersburgo, Krasnodar, Ekaterimburgo]
# Lista de nombres de ciudades
nombres_ciudades = ['Moscu', 'Kaliningrado', 'Petersburgo', 'Krasnodar', 'Ekaterimburgo']

# Iterar sobre cada ciudad y procesar sus datos
for i in range(len(ciudades_data)):
    process_city_data(ciudades_data[i], nombres_ciudades[i], fechas, años_unicos, fechas_años_unicos)
```
